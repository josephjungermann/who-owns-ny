```python
# import pandas
import pandas as pd
```

```python
# read csv in
df = pd.read_csv('../data/WARN_05072020.csv')
df
```

```python
# Dataset from 01/01/20 - 05/07/20 
# For updated totals, please use 'scrape-all-years.ipynb' in this folder to create/import new CSV
df.head()
```

```python
# reveal each column header
df.info()
```

```python
# Fill all null values with 0 and clean all non-perfect integers in 'affected' column
df = df.fillna(0)
df.affected[df.affected == 'To'] = 0
df.affected[df.affected == '1,004'] = 1004
df.affected[df.affected == '1,166'] = 1166
df.affected[df.affected == '(62'] = 62
df.affected[df.affected == '1,321'] = 1321
```

```python
# Clean total_employees section
df.total_employees[df.total_employees == '----'] = 0
```

```python
# Change 'affected' and 'total_employees' section from string objects to integers 
df['affected'] = df['affected'].astype(int)
df['total_employees'] = df['total_employees'].astype(int)
```

```python
# Check to make sure changes are correct
df.info()
```

```python
df['affected'] = pd.to_numeric(df['affected'])
df.info()
```

```python
df.head()
```

```python
# Find out the type of businesses that were affected the most
df['business_type'].value_counts().reset_index()
```

```python
# Find out the reasons why
df['dislocation'].value_counts()
```

```python
# Clean all Covid-19 related reasons into one group
df.loc[df['dislocation'].str.contains('COVID-19'), 'dislocation'] = 'COVID-19'
dislocation = df['dislocation'].value_counts().reset_index().rename(columns={'index': 'dislocation', 'dislocation': 'count'})
dislocation
```

```python
df['affected'].sum()
```

```python
# Make new list an target only businesses affected by COVID-19
covid_dislocation = "COVID-19"
```

```python
covid_list = df[df.dislocation == covid_dislocation]
```

```python
covid_list
```

```python
# export dataset to CSV.
covid_list.to_csv('../output/WARN_COVID_05072020.csv', index=False)
```

```python
# Discover how many people were issued WARN notices due to COVID-19 by business type/industry. Search for top 20 with .head() code.
affected_industries = covid_list.groupby(['business_type'])['affected'].agg('sum').sort_values(ascending = False).reset_index()
affected_industries
```

```python
# export dataset to CSV.
affected_industries.to_csv('../output/Total_COVID_Affected_by_industry.csv', index=False)
```

```python
#which comapnies lost the most people due to COVID. Search top 20.
covid_list.groupby(['company'])['affected'].agg('sum').sort_values(ascending = False).reset_index().head(20)
```

```python
# Next, track amount of people affected by COVID-19 WARN notices by date
covid_affected = covid_list.groupby(['notice_date'])['affected'].agg('sum').sort_values(ascending = False).reset_index()
covid_affected
```

```python
# export dataset to CSV.
covid_affected.to_csv('../output/Total_COVID_Affected_by_date.csv', index=False)
```

```python
# The amount of WARN notices filed by date due to COVID
pd.set_option('display.max_rows', 50)
date = covid_list['notice_date'].value_counts().reset_index().rename(columns = {'index': 'date', 'notice_date': 'count'})
date = date[~date['date'].str.contains('2019')]
date
```
